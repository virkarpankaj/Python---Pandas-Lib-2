## Apply Function

```python
student_details = pd.read_csv('student-mat.csv')
student_details.head()
```

```python
# adding 2 to each element in column MEDU

student_details['Medu']+=2
student_details.head()
```

```python
# by apply Method - applying to above data so 6+2 ...

student_details['Medu'] = student_details['Medu'].apply(lambda x:x+2)
student_details.head()
```

### Create a new column which returns 1 if age>=18 else return 0.

```python
student_details['Eligibility_criteria'] = student_details['age'].apply(lambda x: 1 if x >=18 else 0)
student_details.head()
```

```python
# def a function   - lambda fufnction is more memory efficient

def get_age(x):
    if x>=18:
        return 1
    else:
        return 0
```

```python
student_details['Eligibility_criteria2'] = student_details['age'].apply(get_age)
student_details.head()
```

#### Merge

*pd.merge(
    left: 'DataFrame | Series',
    right: 'DataFrame | Series',
    how: 'str' = 'inner',
    on: 'IndexLabel | None' = None,
    left_on: 'IndexLabel | None' = None,
    right_on: 'IndexLabel | None' = None,
    left_index: 'bool' = False,
    right_index: 'bool' = False,
    sort: 'bool' = False,
    suffixes: 'Suffixes' = ('_x', '_y'),
    copy: 'bool' = True,
    indicator: 'bool' = False,
    validate: 'str | None' = None,
) -> 'DataFrame'

*Docstring:  Merge DataFrame or named Series objects with a database-style join.

* A named Series object is treated as a DataFrame with a single named column.

```python
m_df1 = pd.DataFrame({"key": ["K0", "K1", "K2", "K3"],
                        "A": ["A0", "A1", "A2", "A3"],
                        "B": ["B0", "B1", "B2", "B3"]})

m_df2 = pd.DataFrame({ "key": ["K0", "K1", "K2", "K3"],
                        "C": ["C0", "C1", "C2", "C3"],
                        "D": ["D0", "D1", "D2", "D3"]})

mrg = pd.merge(left= m_df1,right=m_df2)
mrg
```

```python
mrg1 = pd.merge(left= m_df1,right=m_df2,on='key')
mrg1
```

```python
left = pd.DataFrame({"key1": ["K0", "K0", "K1", "K2"],
                     "key2": ["K0", "K1", "K0", "K1"],
                       "A": ["A0", "A1", "A2", "A3"],
                       "B": ["B0", "B1", "B2", "B3"]})

right = pd.DataFrame( {"key1": ["K0", "K1", "K1", "K2"],
                       "key2": ["K0", "K0", "K0", "K0"],
                          "C": ["C0", "C1", "C2", "C3"],
                          "D": ["D0", "D1", "D2", "D3"]})

print('left is \n',left)
print('right is \n',right)

result = pd.merge(left, right, on=["key1", "key2"])
result
```

```python
result_h = pd.merge(left, right,how='left',on=["key1", "key2"])
result_h
```

```python
result_r = pd.merge(left, right,how='right',on=["key1", "key2"])
result_r
```

```python
result_o = pd.merge(left, right,how='outer',on=["key1", "key2"])
result_o
```

```python
result_cross = pd.merge(left, right,how='cross')
result_cross
```

```python
# duplicate index

left = pd.DataFrame({"A": [1, 2], "B": [2, 2]})

right = pd.DataFrame({"A": [4, 5, 6], "B": [2, 2, 2]})

result = pd.merge(left, right, on="B", how="outer")
result
```

## Grouping Vs Crosstab Vs Pivot-table 

```python
insurance_data = pd.read_csv('insurance.csv')
insurance_data
```

#### Find charges based on region 

```python
insurance_data.groupby(by= 'region')['charges'].mean().round().sort_values(ascending=False)
```

#### Find charges based on region & smoker  - BY GROUP BY

```python
insurance_data.groupby(by = ['region','smoker'])['charges'].mean().round()
```

#### Find charges based on region & smoker  - BY PIVOT TABLE

```python
pd.pivot_table(data= insurance_data, values='charges', index='region',columns = 'smoker',aggfunc='mean')
```

```python
pd.pivot_table(data= insurance_data, index='smoker',columns='region',values = 'charges',aggfunc='mean')
```

```python
## pivot table gives data in DataFrame - Tabuler Form
```

#### Find charges based on region & smoker  - BY CROSSTAB

** Compute a simple cross tabulation of two (or more) factors. By default computes a frequency table of the factors unless an array of values and an aggregation function are passed.

```python
pd.crosstab(index=insurance_data['smoker'],columns=insurance_data['region'])
```

```python
pd.crosstab(index=insurance_data['smoker'],columns=insurance_data['region'],margins=True)
```

```python
pd.crosstab(index=insurance_data['smoker'],columns=insurance_data['region'],values=insurance_data['charges'],aggfunc='mean')
```
